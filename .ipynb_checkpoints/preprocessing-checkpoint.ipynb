```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('sequence.csv')
df.head(22)
```

```python
df_sample=df[(df['time']>=0)&(df['time']<=50)].copy()
df_sample.head()
```

```python
att_team=df[df['team_id']==2]
def_team=df[df['team_id']==1]
```

```python
attacking_xy=[]
defenders_xy=[]
time=[]
for i in range (1,int(df.time.max()+1)):
        attack_combined= np.column_stack((att_team[att_team['time']==i]['x'],att_team[att_team['time']==i]['y']))
        attacking_xy.append(attack_combined)
        defend_combined= np.column_stack((def_team[def_team['time']==i]['x'],def_team[def_team['time']==i]['y']))
        defenders_xy.append(defend_combined)
        time.append(i)
df_att_def=pd.DataFrame()
df_att_def['attackers_xy']=attacking_xy
df_att_def['defenders_xy']=defenders_xy
df_att_def['time']=time
```

```python
from scipy.spatial import distance
dist_o_defender_lop=[]
dist_op_defender_lop=[]

count=0
for index,value in df_att_def.iterrows():
    attackers_xy= value[['attackers_xy']].values.tolist()#zip(value['home_team_x'],value['home_team_y'])#value[['home_team_x','home_team_y']].values
    defender_xy=(value[['defenders_xy']]).values.tolist()
    
#     print len(defender_xy[0][10])
    dist_o_defender_lop=[]
    
    for i in range(0,len(attackers_xy[0])):  
    
        dist_defender_lop=[] 
        for j in range(0,len(defender_xy[0])):
            
            lop= distance.euclidean(attackers_xy[0][i], defender_xy[0][j])
            dist_defender_lop.append(lop)
        
        dist_o_defender_lop.append(np.round(min(dist_defender_lop),2))
    dist_op_defender_lop.append(dist_o_defender_lop)
df_att_def['min_distance_att_def']=dist_op_defender_lop
```
